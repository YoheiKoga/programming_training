```python
import statsmodels.datasets as sd
import numpy as np

data = sd.longley.load_pandas()
%load_ext rpy2.ipython
```

```python
data.endog_name, data.exog_name
y, x = data.endog, data.exog

# for convenience, add exogenous data to DataFrame x
x['TOTEMP'] = y
# x
```

```python
gnp = x['GNP']
totemp = x['TOTEMP']
%R -i totemp,gnp 
%R plot(gnp, totemp)
```

```python
xxx = np.arange(1,101,1)
yyy = np.random.randn(100)
yyy = yyy + xxx/10

%R -i xxx,yyy plot(xxx, yyy)
```

```python
%%R
# least-squares regression
fit <- lm(totemp~gnp)

# display the coefficients of the fit
print(fit$coefficients)
# plot the data points
plot(gnp, totemp)
# and plot the linear regression
abline(fit)
```
