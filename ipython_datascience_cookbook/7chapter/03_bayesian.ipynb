```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

posterior = lambda n, h, q:((n+1) * st.binom(n, q).pmf(h))

n = 100
h = 61
q = np.linspace(0., 1., 1000)
d = posterior(n, h, q)
# d2 = (n+1)*st.binom(n, q).pmf(h)
```
