```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
%matplotlib inline
%matplotlib nbagg

custom_style = {'axes.labelcolor': 'white',
                'xtick.color': 'white',
                'ytick.color': 'white'}
sns.set_style("darkgrid", rc=custom_style)
```

```python
fig = plt.figure()
x = np.arange(0, 10, 0.1)

ims = []
for a in range(50):
    y = np.sin(x - a)
    im = plt.plot(x, y, "r")
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims)
```

```python
def update_radius(i, circle):
    circle.radius = i*0.5
    return circle,

fig = plt.gcf()
ax = plt.axes(xlim=(-10,10), ylim=(-10,10))
ax.set_aspect('equal')
circle = plt.Circle((0,0), 0.05)
ax.add_patch(circle)
anim = animation.FuncAnimation(fig, update_radius, fargs=(circle, ), frames=30, interval=10)
anim
```
