```python
# [decision_tree_CART](http://localhost:8888/notebooks/ipython_datascience_cookbook/test/decision_tree_CART.ipynb)

from sklearn.datasets import load_iris
from sklearn import tree
from IPython import display
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
import pydotplus
%matplotlib inline
```

```python
iris = load_iris()

# 説明変数
iris.data[0]
# 目的変数
iris.target

clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(iris.data, iris.target)
```

```python
predicted = clf.predict(iris.data)

sum(predicted == iris.target) / len(iris.target)
```

```python
# export_graphvizでdotファイルとしてエクスポート、pydotplusでPDFファイルに変換して可視化

export_graphviz(clf, out_file="tree.dot",
                feature_names=iris.feature_names,
                class_names=iris.target_names,
                filled=True, rounded=True)
```

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=iris.feature_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
# graph.write_pdf("graph.pdf")
```

```python
# [Decision tree – Python](https://pythonprogramminglanguage.com/decision-tree/)
from sklearn import tree
from IPython import display
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
import pydotplus
import numpy as np
%matplotlib inline
```

```python
# [height, hair-length, voice-pitch]
# X = [[180, 15, 0],
#      [167, 42, 1],
#      [136, 35, 1],
#      [174, 15, 0],
#      [141, 28, 1]
#     ]
X = np.array([[180, 15, 0, 70],
     [167, 42, 1, 56],
     [136, 35, 1, 45],
     [174, 15, 0, 64],
     [141, 28, 1, 46]
    ])
Y = ['man', 'woman', 'woman', 'man', 'woman']

clf = tree.DecisionTreeClassifier()
clf.fit(X, Y)
prediction = clf.predict([[133, 37, 1, 40]])
print(prediction)

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=["height", "hair-length", "voice-pitch", "weight"])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

display.Image(graph.create_png())
```
