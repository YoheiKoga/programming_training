```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://github.com/ipython-books/cookbook-data/raw/master/bikes.csv"

df = pd.read_csv(url, index_col='Date',
                                parse_dates=True, dayfirst=True)
df.head(2)
```

```python
df.describe()
```

```python
# df[['Berri1', 'PierDup']].plot(figsize=(10,6))
df[['Berri1', 'PierDup']].plot()
```

```python
df.index.weekday
```

```python
days = np.array(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
df['Weekday'] = days[df.index.weekday]
```

```python
df_week = df.groupby('Weekday').sum()
df_week
```

```python
df_week.ix[days].plot(lw=3, figsize=(10,6))
plt.ylim(0);
```

```python
from ipywidgets import interact
@interact
def plot(n=(1,30)):
    #pd.rolling_mean(df['Berri1'], n).dropna().plot()
    # pd.rolling(df['Berri1'], n).mean().dropna().plot()
    (df['Berri1']).rolling(window=n, center=False).mean().dropna().plot()
    plt.ylim(0, 8000)
    plt.show()
```
