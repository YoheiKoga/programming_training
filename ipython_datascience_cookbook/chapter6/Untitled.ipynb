```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(16,4))

ax = fig.add_subplot(1,3,1)
ax.set_title('y = x')
x = np.linspace(-5, 5, 100)
ax.plot(x, x)

ax = fig.add_subplot(1, 3, 2)
ax.set_title('y = x^2')
x = np.linspace(-5, 5, 100)
ax.plot(x, x**2)
```
