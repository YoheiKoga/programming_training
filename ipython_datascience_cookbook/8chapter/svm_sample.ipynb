```python
# [Python Programming Tutorials](https://pythonprogramming.net/linear-svc-example-scikit-learn-svm-python/)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn import svm

```

```python
x = [1,5,1.5,8,1,9]
y = [2,8,1.8,8,0.6,11]

plt.scatter(x, y)
plt.show()
```

```python
X = np.array([[1,2],
              [5,8],
              [1.5, 1.8],
              [8,8],
              [1,0.6],
              [9,11]])
y = [0,1,0,1,0,1]

# clf: classifier
# C = 1.0: valuation of "how badly" you want to properly classify, or fit, everything.
# C is cost parameter?
clf = svm.SVC(kernel='linear', C = 1.0)
clf.fit(X,y)
```

```python
clf.predict([[0.58, 0.76],[10, 10],[5,5],[4.3,4.3], [9,1], [1,9]])
```

```python
w = clf.coef_[0]

a = -w[0] / w[1]

xx = np.linspace(0,12)
yy = a * xx - clf.intercept_[0]/w[1]

plt.figure(figsize=(10,10))
h0 = plt.plot(xx, yy, 'k-', label="non weighted div")

plt.scatter(X[:,0], X[:,1], c = ["blue", "green"])
plt.legend()
plt.show()
```
