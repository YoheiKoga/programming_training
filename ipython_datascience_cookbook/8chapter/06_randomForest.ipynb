```python
import numpy as np
import sklearn as sk
import sklearn.datasets as skd
import sklearn.ensemble as ske
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline


# mpl.rcParams['figure.dpi'] = mpl.rcParams['savefig.dpi'] = 300

custom_style = {'axes.labelcolor': 'black',
                'xtick.color': 'black',
                'ytick.color': 'black'}
sns.set_style("darkgrid", rc=custom_style)
```

```python
data = skd.load_boston()

# create RandomforestRegressor model
reg = ske.RandomForestRegressor()

X = data['data']
y = data['target']

reg.fit(X, y)

# importance of features can be found in reg.feature_importances_. Sort them by  descreasing order of importance
fet_ind = np.argsort(reg.feature_importances_)[::-1]
fet_imp = reg.feature_importances_[fet_ind]

# plot histogram
fig = plt.figure(figsize=(16,8))
ax = plt.subplot(111)
# plt.hist(fet_imp)
# plt.bar(np.arange(len(fet_imp)), fet_imp, lw=1, width=0.5, align="edge")
plt.bar(np.arange(len(fet_imp)), fet_imp, lw=1, width=0.5)
plt.grid(False)
ax.set_xticks(np.arange(len(fet_imp)))
ax.set_xticklabels(data['feature_names'][fet_ind])
plt.xlim(-1, len(fet_imp))
```
